```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
myList = []

url = "https://www.accessdata.fda.gov/scripts/cdrh/cfdocs/cfPMN/pmn.cfm"

# Headers to simulate a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
    'Connection': 'keep-alive',
}

# Start a session to handle cookies and headers
session = requests.Session()

# Send a GET request to the page with headers
response = session.get(url, headers=headers, allow_redirects=True)

# Check if there were too many redirects
if response.status_code == 200:
    print("Successfully fetched the page!")
    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extracting all paragraphs from the page
    for paragraph in soup.find_all('p'):
        myList.append(
            {"Type": "Paragraph", "Content": paragraph.get_text().strip()})

    # Extracting headlines if they are inside <h1>, <h2>, etc.
    for headline in soup.find_all(['h1', 'h2']):
        myList.append(
            {"Type": "headline", "Content": headline.get_text().strip()})

    for classes in soup.find_all(class_='pmn'):
        myList.append({"Type": "Class", "Content": classes.get_text().strip()})
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
    print(f"Headers: {response.headers}")


df = pd.DataFrame(myList)
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

# Set up Chrome WebDriver with the Service class
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.get('https://www.geeksforgeeks.org/courses/category/all')


myList = []
try:
    elementsByClass = WebDriverWait(driver, 10).until(
        # EC.visibility_of_all_elements_located((By.CLASS_NAME, 'ui container'))
        EC.visibility_of_all_elements_located(
            (By.XPATH, '//*[contains(@class, "ui cards courseListingPage_cardLayout__multW courseListingPage_courseCardsGrid__VYBzZ")]'))
    )
    print("\nData from Class:")
    for el in elementsByClass:
        myList.append({"courses": el.text})
        # print(el.text)
except Exception as e:
    print(f"Error: {e}")
    print("Elements with the specified class not found or not visible!")

# Close the browser
driver.quit()

# Convert the data into a pandas DataFrame
data = {"Courses": myList}
df = pd.DataFrame(data).to_csv("/home/ankush/Downloads/selenium.csv")
print(df)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Initialize lists to store the data
categories = []
types = []
descriptions = []

# Define the URL and headers for the request
url = "https://i4c.mha.gov.in/cyber-crime-categories.aspx#collapseTwo"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
    'Connection': 'keep-alive',
}

# Send GET request
response = requests.get(url, headers=headers, allow_redirects=True)

# Parse the HTML response
soup = BeautifulSoup(response.text, 'html.parser')

# Find the section-title (categories)
categories_section = soup.find_all('div', class_='section-title')
for category in categories_section:
    categories.append(category.text.strip())  # Extract and clean category name

# Find the accordion headers (types)
heading = soup.find_all('h2', class_='accordion-header')
for header in heading:
    types.append(header.text.strip())  # Extract and clean type name

# Find the data section (descriptions)
data = soup.find_all('div', class_='col-12 col-sm-12 col-md-8')
for description in data:
    # Extract and clean description
    descriptions.append(description.text.strip())

# Ensure the lists are of the same length before creating the DataFrame
min_length = min(len(categories), len(types), len(descriptions))
categories = categories[:min_length]
types = types[:min_length]
descriptions = descriptions[:min_length]

# Create a DataFrame from the lists
df = pd.DataFrame({
    'Category': categories,
    'Type': types,
    'Description': descriptions
}).to_csv('new.csv')

# Display the DataFrame
print(df)
```
